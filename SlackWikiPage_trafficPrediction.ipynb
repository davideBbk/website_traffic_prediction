## 2.Creation of a lagged dataset

```python
views_vals = views['allPlt_allAg']
df_model = pd.concat([views_vals.shift(1), views_vals], axis = 1)

df_model.columns = ['Previous_Step', 'Next_Step']

df_model.head()
```

```python
pd.plotting.autocorrelation_plot(views['allPlt_allAg'])
```

```python
df_model.corr()
```

Looking at autocorrelation plot and correlation of the lagged dataset, a positive high correlation of 0.55 makes possible to deal with the time series problem as an autoregression problem.

## Autoregression

I will keep the first 66% of the observations for training and the remaining 34% for evaluation, excluding the first row of data during the split (it contains a NaN value because there's no previous step for it).
Each of the train and test sets are then split into the input and output variables.

```python
train_size = int(len(views_vals.index.values)*0.66)
#training dataset
train_df = views_vals[1:train_size]

#test dataset
test_df = views_vals[train_size:]

#AR MODEL
ar_model = AR(train_df)
ar_model_fit = ar_model.fit()
print('Lag: %s' % ar_model_fit.k_ar)
print('Coefficients: %s' % ar_model_fit.params)

#PREDICTIONS
predictions = ar_model_fit.predict(start = len(train_df), end = len(train_df) + len(test_df)-1, dynamic = True)
error = mean_squared_error(test_df, predictions)
print('Test MSE: %.3f' % error)
```

```python
for i in range(len(predictions)):
	print('predicted=%f, expected=%f' % (predictions[i], test_df[i]))
```
