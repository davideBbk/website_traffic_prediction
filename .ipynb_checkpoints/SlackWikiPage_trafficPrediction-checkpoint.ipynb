## 2.Creation of a lagged dataset

Creating a lagged dataset I'll be able to make autocorrelation studies and to proceed with an autoregression model at the next step.

```python
views_vals = views['allPlt_allAg']

df_model = pd.concat([views_vals.shift(1), views_vals], axis = 1)

df_model.columns = ['Previous_Step', 'Next_Step']

df_model.head()
```

```python
pd.plotting.autocorrelation_plot(views['allPlt_allAg'])
```

```python
df_model.corr()
```

Looking at autocorrelation plot and correlation of the lagged dataset, a positive high correlation of 0.55 makes possible to deal with the time series problem as an autoregression problem.

## 3.Autoregression

I will keep the first 95% of the observations for training and the remaining portion for evaluation, excluding the first row of data during the split (it contains a NaN value because there's no previous step for it).
Each of the train and test sets are then split into the input and output variables.

```python
train_size = int(len(views_vals.index.values)*0.95)
#training dataset
train_df = views_vals[1:train_size]

#test dataset
test_df = views_vals[train_size:]

#AR MODEL
ar_model = AR(train_df)
ar_model_fit = ar_model.fit()
print('Lag: %s' % ar_model_fit.k_ar)
print('Coefficients: %s' % ar_model_fit.params)

#PREDICTIONS
predictions = ar_model_fit.predict(start = len(train_df), end = len(train_df) + len(test_df)-1, dynamic = True)
error = mean_squared_error(test_df, predictions)
print('Test MSE: %.3f' % error)
```

```python
plt.plot(pd.to_datetime(test_df.index),test_df.values)
plt.plot(predictions)
plt.show()
```

The lag value will be useful for the ARIMA model at the next step.

## 4. ARIMA Forecasting

In order to speed up the analysis, I will use a reduced dataset for training and testing.
As a p parameter for the ARIMA model, I will use the Lag 21 obtained from the previous AR model.

```python
#ARIMA MODEL

new_train = allViews['2016-06':'2017']
new_test = allViews['2018-01':'2018-06']

arima_model = ARIMA(new_train, order = (21,1,0))
arima_model_fit = arima_model.fit()
arima_model_fit.summary()
```

### Plot of residual error

```python
residuals = pd.DataFrame(arima_model_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
residuals.describe()
```

```python
#PREDICTIONS with test data

predictions = []
history = [x for x in new_train]

for t in range(len(new_test)):
	model = ARIMA(history, order=(21,1,0))
	model_fit = model.fit(disp=0)
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = new_test[t]
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
    
```

```python
error = mean_squared_error(new_test, predictions)
print('Test MSE: %.3f' % error)
```

```python
# plot
plt.plot(pd.to_datetime(new_test.index),new_test.values)
plt.plot(pd.to_datetime(new_test.index),predictions, color='red')
plt.show()

```

## 5.Forecast

```python
#Prediction

predictions_ny = []

numdays = 1

for t_1 in range(numdays): 
    model = ARIMA(history, order = (21,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions_ny.append(yhat)
```

```python

base = pd.to_datetime(allViews.index[-1])
date_list = [base - datetime.timedelta(days=x) for x in range(0, numdays)]

plt.plot(pd.to_datetime(allViews.index),allViews.values)
plt.plot(date_list,predictions_ny, color='red')
plt.show()
```
